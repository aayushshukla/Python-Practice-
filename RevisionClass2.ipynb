```python
l=[10,11,33,32,45,52,61,71,22]
# wap to print all even number in a list  and all odd numbers list  from this list with loop and without loop
oddList=[]
evenList=[]
for item in l:
  if item%2==0:evenList.append(item)
  else: oddList.append(item)
print(evenList)
print(oddList)


```

```python
#filter(function,iterable) -- returns filter object
def fn(item):
  if item%2==0:
    return item
evenFilter =list(filter(fn,l))
print(evenFilter)
```

```python
def fn(item):
  if item%2!=0:
    return item
oddFilter =list(filter(fn,l))
print(oddFilter)
```

```python
evenFilter =list( filter(lambda x: x%2==0,l))
oddFilter =list( filter(lambda x: x%2!=0,l))
print(evenFilter)
print(oddFilter)
```

```python
l=['a','e','x','y','n','o','j','u','r','i','k']
# wap to print all the vowels a , e , i , o , u
vowel=['a','e','i','o','u']
vlist= list(filter(lambda item : item in vowel ,l))
print(vlist)
```
